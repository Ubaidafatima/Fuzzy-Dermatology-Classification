```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
data = pd.read_excel('Dermatology dataset.xlsx')

# Disease Labels
disease_names = {
    1: 'Psoriasis',
    2: 'Seborrheic\nDermatitis',
    3: 'Lichen\nPlanus',
    4: 'Chronic\nDermatitis',
    5: 'Pityriasis\nRosea',
    6: 'Pityriasis\nRubra Pilaris'
}

# Create Plot
plt.figure(figsize=(14,7), dpi=300)

colors = sns.color_palette("Set2")

ax = sns.countplot(
    x=data.iloc[:, -1],
    palette=colors
)

# Replace x labels
ax.set_xticklabels(
    [disease_names[i] for i in sorted(disease_names.keys())],
    rotation=0
)

# Title and labels
plt.title(
    'Distribution of Dermatological Disease Classes',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel(
    'Dermatological Disease Classes',
    fontsize=14
)

plt.ylabel(
    'Number of Patients',
    fontsize=14
)

# Add values on bars
for p in ax.patches:
    ax.annotate(
        f'{int(p.get_height())}',
        (p.get_x() + p.get_width()/2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=11
    )

# Grid
plt.grid(axis='y', linestyle='--', alpha=0.4)

# Layout
plt.tight_layout()

# Save figure
plt.savefig(
    'Dermatology_Class_Distribution_HD.png',
    dpi=300,
    bbox_inches='tight'
)

plt.show()
```

```python
# Clinical Symptoms
fuzzy_data['P1'] = fuzzy_data[
    ['erythema',
     'scaling',
     'itching',
     'definite_borders']
].mean(axis=1)

# Epidermal Changes
fuzzy_data['P2'] = fuzzy_data[
    ['acanthosis',
     'hyperkeratosis',
     'parakeratosis',
     'clubbing_rete_ridges']
].mean(axis=1)

# Inflammatory Features
fuzzy_data['P3'] = fuzzy_data[
    ['PNL_infiltrate',
     'eosinophils_infiltrate',
     'exocytosis',
     'spongiosis']
].mean(axis=1)

# Histopathological Features
fuzzy_data['P4'] = fuzzy_data[
    ['elongation_rete_ridges',
     'thinning_suprapapillary_epidermis',
     'munro_microabcess',
     'spongiform_pustule']
].mean(axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import *

import time

X_fuzzy = fuzzy_final[['P1','P2','P3','P4']]
y_fuzzy = fuzzy_final['class']

X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(
    X_fuzzy,
    y_fuzzy,
    test_size=0.2,
    random_state=42,
    stratify=y_fuzzy
)

rf_fuzzy = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

start = time.time()

rf_fuzzy.fit(X_train_f,y_train_f)

end = time.time()

fuzzy_time = end - start

y_pred_f = rf_fuzzy.predict(X_test_f)

print("Accuracy:",
      accuracy_score(y_test_f,y_pred_f))

print("Precision:",
      precision_score(y_test_f,y_pred_f,
                      average='weighted'))

print("Recall:",
      recall_score(y_test_f,y_pred_f,
                   average='weighted'))

print("F1:",
      f1_score(y_test_f,y_pred_f,
               average='weighted'))

print("Kappa:",
      cohen_kappa_score(y_test_f,y_pred_f))

print("Training Time:",
      fuzzy_time)
```

```python
from statsmodels.stats.contingency_tables import mcnemar
import pandas as pd
```

```python
def mcnemar_compare(y_true, pred1, pred2, model1, model2):

    both_correct = 0
    only_model1 = 0
    only_model2 = 0
    both_wrong = 0

    for yt, p1, p2 in zip(y_true, pred1, pred2):

        c1 = (p1 == yt)
        c2 = (p2 == yt)

        if c1 and c2:
            both_correct += 1
        elif c1 and not c2:
            only_model1 += 1
        elif not c1 and c2:
            only_model2 += 1
        else:
            both_wrong += 1

    table = [
        [both_correct, only_model1],
        [only_model2, both_wrong]
    ]

    result = mcnemar(
        table,
        exact=False,
        correction=True
    )

    print("\n" + "="*60)
    print(f"{model1} vs {model2}")
    print("="*60)

    print(pd.DataFrame(
        table,
        index=[f"{model1} Correct", f"{model1} Wrong"],
        columns=[f"{model2} Correct", f"{model2} Wrong"]
    ))

    print("\nChi-square =", round(result.statistic,4))
    print("P-value    =", round(result.pvalue,6))

    if result.pvalue < 0.05:
        print("Result: Statistically Significant")
    else:
        print("Result: Not Statistically Significant")

    return result
```
